```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
data=pd.read_csv("groceries.csv",header=None,error_bad_lines=False)
```

```python
data.head(100)
```

```python
data.shape
```

```python
record=[]
for i in range(0,6106):
    record.append([str(data.values[i,j]) for j in range(0, 4)])
        
```

```python
data.values[0,0]
```

```python
print(record)
```

```python
for i in record:
    print(i)
```

```python
association_rules = apriori(record, min_support = 0.5, min_confidence = .7, min_length = 2)
```

```python
print(association_rules)
```

```python
ass=list(association_rules)
print(len(ass))
```

```python
print(ass)
```

```python
association_rules1 = apriori(record, min_support = 0.0055, min_confidence = .3, min_lift = 3, min_length = 2)
```

```python
ass1=list(association_rules1)
print(len(ass1))
```

```python
print(ass1)
```

```python

for item in ass1:
    pair = item[0]
    items = [x for x in pair]
    print("Rule :"+ str(items[0]) + "->" + str(items[1]))
    print("Support : {}".format(item[1]))
    print("Confidence : {}".format(item[2][0][2]))
    print("List : {}".format(item[2][0][3]))
    print("\n-------------------------------------------------\n")
```
